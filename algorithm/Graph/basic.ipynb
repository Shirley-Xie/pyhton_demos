# 207. Course Schedule:BFS
    有n种课程，当你选择一个课程你必须先选择他的前面一个。成对出现。
    [[1,0],[0,1]]是错误的，因为要上1则必须上0课，矛盾。

```python
class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        # 这图是否存在闭环，BFS
        in_degree = [0]*(numCourses)
        # 先建立邻接表
        graph = {i:[] for i in range(numCourses)}
        for i in prerequisites:
            # 前面是入度
            in_degree[i[0]]+=1
            graph[i[1]].append(i[0])  
                
        visited = False*len(graph)
        
        # 若入度为0则入队
        queue = []
        for i in range(numCourses):
            if in_degree[i] == 0:
                queue.append(i)
                
        cnt = 0
        while queue:
            u = queue.pop(0)
            # 遍历删除的值
            for i in graph[u]:
                in_degree[i]-=1
                if in_degree[i] == 0:
                    queue.append(i)
            cnt+=1            
        if cnt != len(graph):
            return False
        else:
            return True
        
inputs = [[1,0],[0,1]]     
a = Solution()        
a.canFinish(2,inputs)        
```

## 210. Course Schedule II
将选择的课程顺顺序返回，若不存在则返回空。

    Input: 4, [[1,0],[2,0],[3,1],[3,2]]
    Output: [0,1,2,3] or [0,2,1,3]

```python
# BFS
def findOrder1(self, numCourses, prerequisites):
    dic = {i: set() for i in range(numCourses)}
    neigh = collections.defaultdict(set)
    for i, j in prerequisites:
        dic[i].add(j)
        neigh[j].add(i)
    # queue stores the courses which have no prerequisites
    queue = collections.deque([i for i in dic if not dic[i]])
    count, res = 0, []
    while queue:
        node = queue.popleft()
        res.append(node)
        count += 1
        for i in neigh[node]:
            dic[i].remove(node)
            if not dic[i]:
                queue.append(i)
    return res if count == numCourses else []
    
# DFS
def findOrder(self, numCourses, prerequisites):
    dic = collections.defaultdict(set)
    neigh = collections.defaultdict(set)
    for i, j in prerequisites:
        dic[i].add(j)
        neigh[j].add(i)
    stack = [i for i in range(numCourses) if not dic[i]]
    res = []
    while stack:
        node = stack.pop()
        res.append(node)
        for i in neigh[node]:
            dic[i].remove(node)
            if not dic[i]:
                stack.append(i)
        dic.pop(node)
    return res if not dic else []
```
