```python
'''
The problem is  :
Given an ARRAY, to find the longest and increasing sub ARRAY in that given ARRAY and return it.
Example: [10, 22, 9, 33, 21, 50, 41, 60, 80] as input will return [10, 22, 33, 41, 60, 80] as output
'''
v = [1,4,3,2,6,5]
# 找不连续的递增序列

def LIS(arr):
    # 最小答案
    dp = [1 for i in range(len(arr))]
    n=len(arr)
    
    # 递推,若后面的大于前面的加一
    for i in range(1, n):
        tmax = 1
        for j in range(0,i):
            if arr[j] <= arr[i]:
                tmax = max(tmax, dp[j]+1)
        dp[i] = max(tmax, dp[i])
    print(dp,'jjjjjjjjj')
# LIS(v)    

# 方法同上
def longestIncreasingSubsequence(nums):
    # dp:当前位置最大值
    dp=[1 for i in range(len(nums))]
    maxresult=0
    for i in range(1,len(nums)):
        for j in range(0,i):
            if(nums[j]<nums[i]):
                # 不是直接dp[i]=dp[j]+1,因为他是一步一步的算，留住最大的值，这样算是只留最后一步且不一定是最大值
                dp[i]=max(dp[i],dp[j]+1)
        # 这步是找到所有最大的值
        maxresult=max(dp[i],maxresult)
    return maxresult

# print(longestIncreasingSubsequence(v))


"""
This programme outputs the Longest Strictly Increasing Subsequence in O(NLogN)
Where N is the Number of elements in the list 
"""
# 二分查找法
def CeilIndex(v,l,r,key):
    while r-l > 1:
        m = (l + r)//2
        if v[m] >= key:
            r = m
        else:
            l = m
    return r

def LongestIncreasingSubsequenceLength(v):
    if(len(v) == 0):
        return 0
    # 全为零的tail
    tail = [0]*len(v)
    length = 1

    tail[0] = v[0]
    for i in range(1,len(v)):
        # 若比第一个值小，则取较小值。
        if v[i] < tail[0]:
            tail[0] = v[i]
        # 若大于大于最后一个值，将合格值付给最后的值
        elif v[i] > tail[length-1]:
            tail[length] = v[i]
            length += 1  
        else:
            tail[CeilIndex(tail,-1,length-1,v[i])] = v[i]
        print('tail',tail)     
    return length
# v = [1,4,3,2,6,5]
print(LongestIncreasingSubsequenceLength(v))
```
